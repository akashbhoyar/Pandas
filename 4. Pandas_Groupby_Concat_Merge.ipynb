### Concat 

```python
india_weather=pd.DataFrame(
                            {'city':['mumbai','delhi','pune'],
                            'temp':[30,35,28],
                            'humidity':[80,72,60]}
                          )
india_weather
```

```python
usa_weather=pd.DataFrame(
                            {'city':['newyork','chicago','orlando'],
                            'temp':[24,18,10],
                            'humidity':[60,49,70]}
                          )
usa_weather
```

```python
df=pd.concat([india_weather,usa_weather]) #ignore_index=False
df
```

```python
#ignore_index
df=pd.concat([india_weather,usa_weather],ignore_index=True)
df
```

```python
df=pd.concat([india_weather,usa_weather],keys=['india','usa'])
df
```

```python
df.loc['india']
```

```python
df.loc['usa']
```

```python
temp_df=pd.DataFrame(
                    {'city':['mumbai','delhi','bang'],
                     'temp':[32,45,30],   
                    },index=[0,1,2]
                    )

temp_df
```

```python
windspeed_df=pd.DataFrame(
                    {'city':['delhi','mumbai'],
                     'wind':[12,15],   
                    },index=[1,0]
                    )

windspeed_df
```

```python
#axis=1 - it will add second dataframe columnwise

df=pd.concat([temp_df,windspeed_df],axis=1)
df
```

```python
df1=pd.concat([temp_df,windspeed_df],axis=0)
df1
```

### Merged

```python
df1=pd.DataFrame(
                {'city':['newyork','chicago','orlando'],
                 'temp':[21,15,18]
                    
                })
df1
```

```python
df2=pd.DataFrame(
                {'city':['chicago','newyork','orlando'],
                 'humidity':[65,70,75]
                    
                })
df2
```

```python
df3=pd.merge(df1,df2,on='city')
df3
```

```python
df1=pd.DataFrame(
                {'city':['newyork','chicago','orlando','baltimore'],
                 'temp':[21,15,18,25]
                    
                })
df1
```

```python
df2=pd.DataFrame(
                {'city':['chicago','newyork','san_diego'],
                 'humidity':[65,70,75]
                    
                })
df2
```

```python
df3=pd.merge(df1,df2,on='city')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='inner')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='outer')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='left')
df3
```
