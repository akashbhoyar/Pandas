```python
#will replace all '-1' with NaN in dataframe

df=pd.read_csv('stock_data.csv',delimiter='\t',na_values=[-1])
df
```

```python
#will replace only '-1' of revenue column with NaN.

df=pd.read_csv('stock_data.csv',delimiter='\t',na_values={'eps':['not available',-1],
                                                          'revenue':[-1],
                                                         'price':'n.a.',
                                                         'people':'n.a.'})
df
```
