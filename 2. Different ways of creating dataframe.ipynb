### creating datafram by using dictionary

```python
weather_data={'day':['18/2/22','19/2/22','20/2/22','21/2/22','22/2/22'],
             'temperature':[30,31,33,29,33],
             'windspeed':[10,9,12,15,18],
             'event':['Rain','Sunny','Snow','Rain','Snow']}
print(weather_data)

df=pd.DataFrame(weather_data)
df
```

### creating the dataframe using list of tuple

```python
weather_data=[('23/02/22',32,6,'Sunny'),('24/02/22',29,9,'humid'),('25/02/22',33,12,'Rainy')]
print(weather_data)
type(weather_data)
```

```python
len(weather_data)
```

```python
weather_data=[('23/02/22',32,6,'Sunny'),('24/02/22',29,9,'humid'),('25/02/22',33,12,'Rainy')]

df=pd.DataFrame(weather_data,columns=['day','temp','wind_speed','event'])
df
```

### creating dataframe using list of dictionary


```python
weather_data=[{'day':'23/02/22','temp':32,'wind_speed':6,'event':'Sunny'},
              {'day':'24/02/22','temp':29,'wind_speed':9,'event':'Humid'},
              {'day':'25/02/22','temp':33,'wind_speed':12,'event':'Rainy'}]

df=pd.DataFrame(weather_data)
df
```
